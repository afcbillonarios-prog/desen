# Fase 1: Identificación del Problema

En esta etapa, identificamos el problema de negocio. El objetivo es analizar los desembolsos de crédito de fomento agropecuario y predecir el **valor del crédito** (`valor_credito`) otorgado, basándonos en variables como el plazo, el periodo de gracia y otras características del crédito.

# Fase 2: Recolección de Datos

Cargamos la muestra de los datos obtenidos del sistema original para comenzar a trabajar (hemos extraído una muestra de 50,000 registros para facilitar el procesamiento).

```python
import pandas as pd
import numpy as np

# Cargar los datos
df = pd.read_csv('sample_data.csv')
df.head()
```

# Fase 3: Preparación de Datos (ETL)

Realizamos la limpieza de los datos, tratamos valores nulos y adaptamos los tipos de datos. También seleccionamos las columnas de interés para nuestro análisis y modelo.

```python
# Convertir valores numéricos removiendo comas
for col in ['valor_credito', 'valor_inversion']:
    if col in df.columns:
        df[col] = df[col].astype(str).replace(r'[^0-9.-]', '', regex=True).astype(float)

# Seleccionar variables numéricas útiles y limpiar nulos
columns_to_keep = ['plazo', 'periodo_gracia', 'cuotas', 'cuotas_capital', 'tasa_indexacion', 'valor_credito']
columns_to_keep = [c for c in columns_to_keep if c in df.columns]
df_clean = df[columns_to_keep].copy()

for col in ['plazo', 'periodo_gracia', 'cuotas', 'cuotas_capital', 'valor_credito']:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

df_clean.dropna(inplace=True)

df_clean.head()
```

```python
# Guardamos los datos limpios para la siguiente etapa
df_clean.to_csv('cleaned_data.csv', index=False)
print('Datos limpios guardados en cleaned_data.csv')
```
